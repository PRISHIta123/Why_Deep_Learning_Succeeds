```python
import tensorflow as tf

tf.debugging.set_log_device_placement(True)

tf.random.set_seed(1234)

from tensorflow.keras import layers, models
from tensorflow.keras.datasets import cifar10, mnist, fashion_mnist
import matplotlib.pyplot as plt
```

```python
def CNN(n_layers):
  CNN=[]

  # Input shape is (28,28,1) for MNIST and Fashion MNIST and (32,32,3) for CIFAR10, change input_shape parameter accordingly
  CNN.append(layers.Conv2D(64, (3, 3), activation='relu', padding='same', input_shape=(28, 28, 1)))

  for _ in range(n_layers-2):
    CNN.append(layers.Conv2D(64, (3, 3), activation='relu', padding='same'))

  CNN.append(layers.Flatten())
  CNN.append(layers.Dense(10))

  return CNN
```

```python
import jax
import jax.numpy as jnp
import numpy as np

# Uncomment required dataset
#(x_train, y_train), (x_valid, y_valid) = fashion_mnist.load_data()
(x_train, y_train), (x_valid, y_valid) = mnist.load_data()
#(x_train, y_train), (x_valid, y_valid) = cifar10.load_data()

# Uncomment for MNIST and Fashion MNIST
x_train= x_train[:,:,:,np.newaxis]
x_valid= x_valid[:,:,:,np.newaxis]
y_train= y_train[:,np.newaxis]
y_valid= y_valid[:,np.newaxis]

# Normalize the image pixels in the range [0, 1]
x_train_normalized = np.array(jnp.array(x_train[:500] / 255.))
x_valid_normalized = np.array(jnp.array(x_valid[:50] / 255.))

y_train_ohe = np.array(jnp.squeeze(jax.nn.one_hot(y_train[:500], num_classes=10)))
y_valid_ohe = np.array(jnp.squeeze(jax.nn.one_hot(y_valid[:50], num_classes=10)))
```

```python
model = models.Sequential()

cnn= CNN(n_layers=5)

for layer in cnn:
  print(layer)
  model.add(layer)

model.summary()
```

```python
optimizer= tf.keras.optimizers.SGD(learning_rate=0.1)

model.compile(optimizer=optimizer,
              loss=tf.keras.losses.MeanSquaredError(),
              metrics=['accuracy'])

model.fit(x_train_normalized[0:500], y_train_ohe[0:500], epochs=50,validation_data=(x_valid_normalized[0:50], y_valid_ohe[0:50]))
```

```python
test_loss, test_acc = model.evaluate(x_valid_normalized[0:50],y_valid_ohe[0:50], verbose=2)
```

```python
from neural_tangents import stax
import neural_tangents as nt

def CNN(n_layers,num_classes=10):

  blocks=[]

  for _ in range(n_layers-1):
    blocks+=[stax.Conv(64, (3, 3),strides=(1,1), padding='SAME'),stax.Relu()]

  blocks+=[stax.Flatten(),stax.Dense(num_classes, 1., 0.)]

  return stax.serial(*blocks)

def CNN_GAP(n_layers,num_classes=10):
  
  blocks=[]

  for _ in range(n_layers-1):
    blocks+=[stax.Conv(64, (3, 3),strides=(1,1), padding='SAME'),stax.Relu()]

  blocks+=[stax.AvgPool((8, 8)),stax.Flatten(),stax.Dense(num_classes, 1., 0.)]

  return stax.serial(*blocks)
```

```python
import tensorflow as tf
from tensorflow.keras.datasets import cifar10, mnist, fashion_mnist
```

```python
import jax
import jax.numpy as jnp
import numpy as np

#Uncomment required dataset
#(x_train, y_train), (x_valid, y_valid) = fashion_mnist.load_data()
(x_train, y_train), (x_valid, y_valid) = mnist.load_data()
#(x_train, y_train), (x_valid, y_valid) = cifar10.load_data()

# Uncomment for MNIST and Fashion MNIST
x_train= x_train[:,:,:,np.newaxis]
x_valid= x_valid[:,:,:,np.newaxis]
y_train= y_train[:,np.newaxis]
y_valid= y_valid[:,np.newaxis]

# Normalize the image pixels in the range [0, 1]
x_train_normalized = jnp.array(x_train[:500] / 255.)
x_valid_normalized = jnp.array(x_valid[:50] / 255.)

# One hot encoding applied to the labels. We have 10
# classes in the dataset, hence the depth of OHE would be 10
y_train_ohe = jnp.squeeze(jax.nn.one_hot(y_train[:500], num_classes=10))
y_valid_ohe = jnp.squeeze(jax.nn.one_hot(y_valid[:50], num_classes=10))
```

```python
#Change n_layers to required number of layers (5,10,12,20)
init_fn, apply_fn, kernel_fn = CNN(n_layers=5,num_classes=10)
#init_fn, apply_fn, kernel_fn = CNN_GAP(n_layers=5,num_classes=10)
```

```python
predict_fn = nt.predict.gradient_descent_mse_ensemble(kernel_fn, x_train_normalized,y_train_ohe)
y_test_ntk = predict_fn(x_test=x_valid_normalized, get='ntk')
```

```python
target_class = jnp.argmax(y_valid_ohe, axis=1)
predicted_class = jnp.argmax(y_test_ntk, axis=1)
acc= jnp.mean(predicted_class == target_class)
print("Accuracy:",acc)
```
